```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("SOCR-HeightWeight.csv")
df
```

```python
df.head()
```

```python
plt.scatter(df['Height(Inches)'], df['Weight(Pounds)'])
plt.xlabel("Height (Inches)")
plt.ylabel("Weight (Pounds)")
plt.title("Height vs Weight Scatter Plot")
plt.show()
```

```python
X=df[["Weight(Pounds)"]]
Y=df[["Height(Inches)"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.25,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_train
X_test
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
print("the slope of the model is:-",regressor.coef_)
print("the intercept of the model is:-",regressor.intercept_)
```

```python
regressor.predict(X_train)
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,regressor.predict(X_train),'r')
```

```python
y_pred_test=regressor.predict(X_test)
y_pred_test
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test,regressor.predict(X_test),'r')
```

```python
from sklearn.metrics import mean_squared_error,r2_score
MSE=mean_squared_error(Y_test,y_pred_test)
MSE
```
