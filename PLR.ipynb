```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
```

```python
X=2*np.random.rand(100,1)
X
```

```python
Y=4+3*X+1.5*X**2+np.random.rand(100,1)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train.shape,X_test.shape,Y_test.shape,Y_train.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features=PolynomialFeatures(degree=2,include_bias=False)
X_poly_train=poly_features.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_poly_train,Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_poly_predict=model.predict(X_poly_train)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_train,y_poly_predict)
```

```python
plt.scatter(X_train,Y_train,label="train_data")
```

```python
plt.scatter(X_train,Y_train,label="train_data")
X_range=np.linspace(0,2,100).reshape(-1,1)
X_range_Poly=poly_features.transform(X_range)
plt.plot(X_range,model.predict(X_range_Poly),color="red",label="Polynomial regression of degree 2")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('PolynomialFeatures')
plt.legend()
plt.show()
```
