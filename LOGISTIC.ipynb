```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('mpg')
df
```

```python
df.drop("name",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['horsepower'].median()
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df["origin"] = df["origin"].astype(str).str.lower().str.strip()

df["origin"] = df["origin"].map({
    "usa": 1,
    "japan": 2,
    "europe": 3
})
```

```python
df["origin"]
```

```python
X=df.drop('mpg',axis=1)
y=df['mpg']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
regressor.coef_
```

```python
for i,col_name in enumerate(X_train.columns):
    print(f"the coefficient for this {col_name} is {regressor.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred_linear=regressor.predict(X_test)
r2_linear=r2_score(Y_test,y_pred_linear)
print("the square of linear regression is ",r2_linear)
```

```python
from sklearn.linear_model import Ridge
ridge_regression_model=Ridge(alpha=0.1)
ridge_regression_model.fit(X_train,Y_train)
```

```python
y_pred_ridge=ridge_regression_model.predict(X_test)
r2_ridge=r2_score(Y_test,y_pred_ridge)
print("the r2_score for this ridge model is ",r2_ridge)
```

```python
for i,col_name in enumerate(X_train.columns):
    print(f"the coefficient for this {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
from sklearn.linear_model import Lasso
Lasso_regression_model=Lasso(alpha=0.1)
Lasso_regression_model.fit(X_train,Y_train)
```

```python
y_pred_Lasso=Lasso_regression_model.predict(X_test)
r2_Lasso=r2_score(Y_test,y_pred_Lasso)
print("the r2_score for this Lasso model is ",r2_Lasso)
```

```python
for i,col_name in enumerate(X_train.columns):
    print(f"the coefficient for this {col_name} is {Lasso_regression_model.coef_[i]}")
```

```python
from sklearn.linear_model import ElasticNet
Elastic_regression_model=ElasticNet(alpha=1,l1_ratio=0.5)
Elastic_regression_model.fit(X_train,Y_train)
```

```python
y_pred_Elastic=Elastic_regression_model.predict(X_test)
r2_Elastic=r2_score(Y_test,y_pred_Elastic)
print("the r2_score for this Lasso model is ",r2_Elastic)
```

```python
for i,col_name in enumerate(X_train.columns):
    print(f"the coefficient for this {col_name} is {Elastic_regression_model.coef_[i]}")
```
